### Dataframes informações diversas

 - Filtrar apenas o último ano disponível no dataframe

```python
#Consumo de álcool per capita 
alcool = pd.read_csv(r'C:\Users\Marcos\Análises\Futebol\total-alcohol-consumption-per-capita-litres-of-pure-alcohol.csv')
al_2018 = alcool[alcool['Year'] == 2018]

#Lista IDH
idh = pd.read_csv(r'C:\Users\Marcos\Análises\Futebol\human-development-index.csv')
idh_2017 = idh[idh['Year'] == 2017]

#Taxa de homicídios - neste caso, o Iran só tem dados até 2014. nivelar por este
hmc = pd.read_csv(r'C:\Users\Marcos\Análises\Futebol\homicide-rate.csv')
hmc_2019 = hmc[hmc['Year'] == 2019]

#Violência no trânsito por 100 mil hab
transito = pd.read_csv(r'C:\Users\Marcos\Análises\Futebol\violencia_transito.csv')
transito_2013 = transito[transito['Year'] == 2013]

#Pib
pib = pd.read_csv(r'C:\Users\Marcos\Análises\Futebol\pib.csv')
pib_2019 = pib[pib['Year'] == 2019]

#Felicidade
#feliz = pd.read_csv(r'C:\Users\Marcos\Análises\Futebol\felicidade.csv')
feliz= pd.read_csv(r'C:\Users\Marcos\Análises\Futebol\feliz.csv')
feliz_2013 = feliz[feliz['Year'] == 2013]
```

### Cruzando os dataframes 

### IDH dos times da Copa

```python
#Cruzando DF de IDH
'''Observação importante: No dataframe dos dados sociais não tem England, apenas UK. Então é preciso
substituir manualmente, trocando o dado de England pelo de Uk,
fazendo o mesmo com Escócia e Gales caso se classifique'''

idh_copa = pd.merge(times, idh_2017, on='country', how='left')
idh_copa.loc[idh_copa['country'] == 'England', 'Human Development Index (UNDP)'] = 922.00
idh_copa.loc[idh_copa['country'] == 'Wales', 'Human Development Index (UNDP)'] = 922.00
idh_times_copa = idh_copa.drop(['Participações', 'Part Consecutivas', 'Última', 'Melhor resultado anterior','Code', 'Year'], axis=1)
idh_times_copa[ 'Human Development Index (UNDP)' ] /= 1000
pd.set_option('display.precision', 2)

```

### Consumo de álcool times da Copa

```python
#Dataframe de quantidade álcool per capita 
'''Observação importante: No dataframe dos dados sociais não tem England, apenas UK. Então é preciso
substituir manualmente, trocando o dado de England pelo de Uk,
fazendo o mesmo com Escócia e Gales caso se classifique'''
alcool_copa = pd.merge(times, al_2018, on='country', how='left')
#abaixo, método para alterar dado diretamente no dataframe
alcool_copa.loc[alcool_copa['country'] == 'England', 'Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)'] = 11.45
alcool_copa.loc[alcool_copa['country'] == 'Wales', 'Total alcohol consumption per capita (liters of pure alcohol, projected estimates, 15+ years of age)'] = 11.45
#pd.set_option('display.precision', 2)
alcool_times_copa = alcool_copa.drop(['Participações', 'Part Consecutivas', 'Última', 'Melhor resultado anterior','Code', 'Year'], axis=1)
```

### Taxa de homicídios times da Copa

```python
#Taxa de homicídios 
homicidios = pd.merge(times, hmc_2019, on='country', how='left' )
homicidios_times_copa = homicidios.drop(['Participações', 'Part Consecutivas', 'Última', 'Melhor resultado anterior','Code', 'Year'], axis=1)
pd.set_option('display.precision', 2)

```

### Números vítimas do trânsito - times da Copa

```python
#Taxa de vítimas fatais no trânsito
vitimas_transito = pd.merge(times, transito_2013, on='country', how='left')
vitimas_transito.loc[vitimas_transito['country'] == 'England', '3.6.1 - Death rate due to road traffic injuries (per 100,000 population) - SH_STA_TRAF'] = 2.9
vitimas_transito.loc[vitimas_transito['country'] == 'Wales', '3.6.1 - Death rate due to road traffic injuries (per 100,000 population) - SH_STA_TRAF'] = 2.9
vitimas_transito_times = vitimas_transito.drop(['Participações', 'Part Consecutivas', 'Última', 'Melhor resultado anterior','Code', 'Year'], axis=1)
pd.set_option('display.precision', 2)
```

### PIB per capita times da Copa

```python
#PIB per capta 
pib_times = pd.merge(times, pib_2019, on='country', how='left')
pib_times.loc[pib_times['country'] == 'England', 'GDP per capita, PPP (constant 2017 international $)'] = 46406.461748
pib_times.loc[pib_times['country'] == 'Wales', 'GDP per capita, PPP (constant 2017 international $)'] = 46406.461748
pib_times_copa = pib_times.drop(['Participações', 'Part Consecutivas', 'Última', 'Melhor resultado anterior','Code', 'Year'], axis=1)
pd.set_option('display.precision', 2)
 
```

### Ranking felicidade times da Copa

```python
### Felicidades (Dados de 2013)
feliz_times = pd.merge(times, feliz_2013, on='country', how='left')
feliz_times.loc[feliz_times['country'] == 'England', 'Life satisfaction in Cantril Ladder (World Happiness Report 2022)'] = 6.515445
feliz_times.loc[feliz_times['country'] == 'Wales', 'Life satisfaction in Cantril Ladder (World Happiness Report 2022)'] = 6.515445
feliz_times_copa = feliz_times.drop(['Participações', 'Part Consecutivas', 'Última', 'Melhor resultado anterior','Code', 'Year'], axis=1)
pd.set_option('display.precision', 2)
```
